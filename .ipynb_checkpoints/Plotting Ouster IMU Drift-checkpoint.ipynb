```python
### Things to do:
# Implement linspace interpolation during GNSS outages
    # Probably wont work (maybe on IMU)
    # https://www.tutorialspoint.com/how-to-plot-a-smooth-line-with-matplotlib
    # https://numpy.org/doc/stable/reference/generated/numpy.linspace.html
# Implement pandas dataframes
    # Reduce many lists into fewer dataframes
    # tidier, faster?

# INGEST DATA AND PLOT INTIAL STATE

# Make sure matplotlib plots inline
%matplotlib inline

import os
import matplotlib.pyplot as plt
import numpy as np
import math
from statistics import pstdev


#Time,GPS_Long,GPS_Lat,GPS_Alt,SDn,SDe,SDu,GPS_Status,IMU_AngVelX,IMU_AngVelY,IMU_AngVelZ,IMU_LinearAccX,IMU_LinearAccY,IMU_LinearAccZ
#0.000014deg (1m difference)
#accelerometer in g
#gyro in deg/s
# Things to remember ^^^^^

# Set and Get directory
os.chdir(r'C:\Users\mikeh\OneDrive\Documents\GitHub\kalmanFilter')
dir = os.getcwd()
print(f'\n\ndirectory: {dir}\n\n')

# Set the input file (full or small)
#infile = 'less_data.txt'
infile = 'C2_IMU.txt'

# Initialize a bunch of empty lists
data = []
imutime, lon, lat, alt, gnss_sd = [], [], [], [], []
status, iavx, iavy, iavz, gnsstime = [], [], [], [], []
ilax, ilay, ilaz, imu_lin_acc, imu_ang_vel = [], [], [], [], []
n_ilax, n_ilay, n_ilaz, sdn, sde = [], [], [], [], []
sdu = []

# Collect all the data from the .txt file and append to appropriate list
with open(infile, 'r') as f:
    f.readline()
    for i, line in enumerate(f):
        data.append(line.replace(' ', '').strip().split(','))
        t, lo, la, al, sd_n, sd_e, sd_u, stat, ia_velx, ia_vely, ia_velz, il_accx, il_accy, il_accz = data[i]
        if i == 0:
            starttime = int(t[11:-1])
        if il_accx != 'None' or il_accz != 'None':
            imutime.append((int(t[11:-1])-starttime)/(10**9))
            ilax.append(float(il_accx))
            ilay.append(float(il_accy))
            ilaz.append(float(il_accz))
            iavx.append(float(ia_velx))
            iavy.append(float(ia_vely))
            iavz.append(float(ia_velz))
            
            normalizer = float(il_accx)**2 + float(il_accy)**2 + float(il_accz)**2
                                
            n_ilax.append(float(il_accx)/math.sqrt(normalizer))
            n_ilay.append(float(il_accy)/math.sqrt(normalizer))
            n_ilaz.append(float(il_accz)/math.sqrt(normalizer))
            
            imu_lin_acc.append([float(il_accx), float(il_accy), float(il_accz)])
            imu_ang_vel.append([float(ia_velx), float(ia_vely), float(ia_velz)])
            
        elif lo != 'None':
            status.append(float(stat))
            gnsstime.append((int(t[11:-1])-starttime)/(10**9))
            lon.append(float(lo))
            lat.append(float(la))
            alt.append(float(al))
            sdn.append(float(sd_n))
            sde.append(float(sd_e))
            sdu.append(float(sd_u))
        
            

# Print Statistics:
print(f'Raw IMU linear acceleration standard deviation \n' + \
        f'\t X: {pstdev(ilax)} \n' + \
        f'\t Y: {pstdev(ilay)} \n' + \
        f'\t Z: {pstdev(ilaz)} \n')

print(f'Raw IMU normalized linear acceleration standard deviation \n' + \
        f'\t nX: \t{pstdev(n_ilax)} \n' + \
        f'\t nY: \t{pstdev(n_ilay)} \n' + \
        f'\t nZ: \t{pstdev(n_ilaz)} \n')

print(f'Raw IMU linear acceleration standard deviation \n' + \
        f'\t X: {pstdev(iavx)} \n' + \
        f'\t Y: {pstdev(iavy)} \n' + \
        f'\t Z: {pstdev(iavz)} \n')

print(imutime[0:10])

# Plot Linear Acceleration          
plt.figure()
plt.title('IMU Linear Acceleration')
plt.plot(imutime, ilax,
        color='red')
plt.plot(imutime, ilay,
        color='green')
plt.plot(imutime, ilaz,
        color='blue')
plt.xlabel('IMU Time (sec)')
plt.ylabel('g (m/s^2)')
plt.yticks([-20, -15, -10, -5, 0, 5, 10, 15, 20])
plt.show()

# Plot Normalized Linear Acceleration
plt.figure()
plt.title('IMU Linear Acceleration (Normalized)')
plt.plot(imutime, n_ilax,
        color='red')
plt.plot(imutime, n_ilay,
        color='green')
plt.plot(imutime, n_ilaz,
        color='blue')
plt.xlabel('IMU Time (sec)')
plt.ylabel('g (m/s^2)')
plt.yticks([-20, -15, -10, -5, 0, 5, 10, 15, 20])
plt.show()

# Plot Angular Velocity
plt.figure()
plt.title('IMU Angular Velocity')
plt.plot(imutime, iavx,
        color='red')
plt.plot(imutime, iavy,
        color='green')
plt.plot(imutime, iavz,
        color='blue')
plt.xlabel('IMU Time (sec)')
plt.ylabel('deg/sec')
plt.show()
            
```

```python
# COMPLEMENTARY FILTER

# Set matplotlib to plot inline
%matplotlib inline

# Set sensor sensitivites
acc_sens = 8192.0
gyro_sens = 16

# Timestep interval (default value)
dt = 0.01 # IMPORTANT time in seconds

# Give them simpler names
acc = imu_lin_acc
gyro = imu_ang_vel

# Initialize Pitch, Roll, and Yaw values
pitch = 0
roll = 0 

# heading = 180 * math.atan2(mag_y, mag_x)/
yaw = 0 # IMPORTANT this is initial heading

# These will be an array of Pitch, Roll, and Yaw values to plot
p = []
r = []
y = []

# Complementary Filter function
def comp_filter(acc, gyro, pitch, roll, yaw, dt): 
    
    # Integrate the gyroscope data -> int(angularSpeed) = angle
    pitch += (gyro[0] / gyro_sens) * dt  # Angle around the X-axis
    roll  += (gyro[1] / gyro_sens) * dt  # Angle around the Y-axis
    yaw   += (gyro[2] / gyro_sens) * dt  # Angle around the Z-axis
    
    
    # (Compensate for drift with accelerometer data)
    # (Sensitivity = -2 to 2 G at 16Bit -> 2G = 32768 && 0.5G = 819)
    # These ^ were removed from this specific implementation
    # Kept for documentation

    # Turning around the X axis results in a vector on the Y-axis
    pitchAcc = math.atan2(acc[1], acc[2]) * (180 / math.pi)
    pitch = pitch * 0.98 + pitchAcc * 0.02

    # Turning around the Y axis results in a vector on the X-axis
    rollAcc = math.atan2(acc[0], acc[2]) * (180 / math.pi)
    roll = roll * 0.98 + rollAcc * 0.02

    # Turning around the Z axis results in a vector on the X-axis 
    yawAcc = math.atan2(acc[0], acc[1]) * (180 / math.pi)
    yaw = yaw * 0.98 + yawAcc * 0.02
    
    out = (pitch, roll, yaw)
    
    return out


for i in range(len(acc)):
    if i == 0:
        dt = imutime[i+1] - imutime[i]
    else:
        dt = imutime[i] - imutime[i-1]
    pitch, roll, yaw = comp_filter(acc[i], gyro[i], pitch, roll, yaw, dt) 
    p.append(pitch)
    r.append(roll+45)
    y.append(yaw+90)


plt.plot(imutime, p,
        color = 'red')
plt.plot(imutime, r,
        color = 'green')
plt.plot(imutime, y,
        color = 'blue')
plt.xlabel('IMU Time (sec)')
plt.ylabel('deg')
plt.show()
```

```python
# CONVERT GNSS DD TO METRES

# Sampled from YeO 
# (https://codereview.stackexchange.com/questions/195933/convert-geodetic-coordinates-to-geocentric-cartesian)

# Ellipsoid Parameters as tuples (semi major axis, inverse flattening)
#grs80 = (6378137, 298.257222100882711)
wgs84 = (6378137, 298.257223563)


def geodetic_to_geocentric(ellps, lat, lon, h):

    # Compute the Geocentric (Cartesian) Coordinates X, Y, Z
    # given the Geodetic Coordinates lat, lon + Ellipsoid Height h
    
    a, rf = ellps
    lat_rad = math.radians(lat)
    lon_rad = math.radians(lon)
    N = a / math.sqrt(1 - (1 - (1 - 1 / rf) ** 2) * (math.sin(lat_rad)) ** 2)
    X = (N + h) * math.cos(lat_rad) * math.cos(lon_rad)
    Y = (N + h) * math.cos(lat_rad) * math.sin(lon_rad)
    Z = ((1 - 1 / rf) ** 2 * N + h) * math.sin(lat_rad)

    return X, Y, Z

new_lat = []
new_lon = []
new_alt = []

reducer = 10000
for i in range(len(lat)):
    tlist = geodetic_to_geocentric(wgs84, lat[i], lon[i], alt[i])
    new_lon.append(tlist[0]-reducer)
    new_lat.append(tlist[1]-reducer)
    new_alt.append(tlist[2]-reducer)
```

```python
# PATH/PLOT GNSS AND IMU DATA
%matplotlib qt

# Instantiate the IMU data
imu_groundx = []
imu_groundy = []
imu_groundz = []

imu_groundx.append(new_lon[0])
imu_groundy.append(new_alt[0])
imu_groundz.append(new_lat[0])

# Gravitational constant
g = -9.805

xheading = 0
yheading = 0 # IMPORTANT will be set by magnetometer

# Calculate a new cartesian value for each IMU time-step
for i in range(len(ilax)-1):
    
    dt = imutime[i+1] - imutime[i]
 
    # Integrate Linear Acc for velocity
    xvel = ilax[i] * dt
    yvel = ilay[i] * dt

    # Get time step (time * inst_velocity)
    x_disp = xvel * dt
    y_disp = yvel * dt
    
    # Get updated headings (both horizontal (yaw) and vertical (pitch))
    
    xheading += y[i+1] - y[i]
    if xheading < 0:
        xheading += 360
    xheading = xheading % 360
    
    yheading += p[i+1] - p[i]
    if yheading < 0:
        yheading += 360
    yheading = yheading % 360
    
    # Iterate 1 step in that heading
    imu_groundx.append(imu_groundx[i] + math.cos(xheading)*x_disp)
    imu_groundy.append(imu_groundy[i] + math.cos(yheading)*y_disp)
    imu_groundz.append(imu_groundz[i] + math.sin(xheading)*x_disp)
     

# Plot the GNSS and IMU tracks
plt.figure()
plt.title("Tracks")
plt.xlabel("longitude")
plt.ylabel("latitude")

plt.scatter(new_lon, new_lat, 
            c=sdn,
            cmap='gnuplot',
            s=status)
plt.plot(imu_groundx, imu_groundz,
            c='red')

plt.show()
```
